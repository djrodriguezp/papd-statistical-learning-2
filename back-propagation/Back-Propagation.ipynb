## Back-Propagation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

Se define la clase NeuralNetworkXOR la cual tiene 1 capa de entrada, 2 capas ocultas y una capa de salida con las siguientes especificaciones:

- Capa de Entrada: 3 neuronas
- Primera Capa Intermedia: 30 neuronas, función de activación: RelU
- Segunda Capa Intermedia 2 neuronas,  función de activación: RelU
- Capa de salida 1 neurona, sin función de activación

```python
class NeuralNetworkXOR:
    def __init__(self,X,Y,lr):
        self.X = X
        self.Y = Y
        self.y_hat = None
        self.h1_neurons = 30
        self.h1_output = None
        self.std = 1
        self.h1_weights = np.random.normal(0, self.std, size=(2,self.h1_neurons))
        self.h1_bias = np.random.normal(0, self.std, size=(1,self.h1_neurons))
        self.h2_output = None
        self.h2_weights = np.random.normal(0, self.std, size=(self.h1_neurons,2))
        self.h2_bias = np.random.normal(0, self.std, 2).reshape(1,2)
        self.y_hat = None
        self.output_weights = np.random.normal(0, self.std, 2).reshape(2,1)
        self.output_bias = np.random.normal(0, self.std, 1).reshape(1,1)
        self.lr = lr
        
        
    def __relu(self,x):
        return x * (x > 0)

    def __relu_prime(self,x):
        return 1 * (x >= 0)
    
    def fit(self):
        for epoch in range(1,10000):
            self.forward_propagation()
            self.back_propagation()
        
    def back_propagation(self):
        error = (self.Y - self.y_hat)
        h2_error_term = error.dot(self.output_weights.T) * self.__relu_prime(self.h2_output)
        h1_error_term = np.matmul(h2_error_term,self.h2_weights.T) * self.__relu_prime(self.h1_output)

        self.output_weights += self.lr * self.h2_output.T.dot(error)
        self.output_bias += self.lr * np.sum(error, axis=0, keepdims=True)
        self.h2_weights += self.lr * self.h1_output.T.dot(h2_error_term)
        self.h2_bias += self.lr * np.sum(h2_error_term, axis=0)
        self.h1_weights += self.lr * self.X.T.dot(h1_error_term)
        self.h1_bias = self.lr * np.sum(h1_error_term, axis=0, keepdims=True)
    
    def forward_propagation(self,X=None):
        if X is None:
            X = self.X
        self.h1_output = self.__relu(X @ self.h1_weights + self.h1_bias) 
        self.h2_output = self.__relu(self.h1_output @ self.h2_weights + self.h2_bias) 
        self.y_hat = (self.h2_output @ self.output_weights  + self.output_bias)
        return self.y_hat
```

Después de varios experimentos, se define un lr=0.001, se procede a entrenar la red 5 veces y se grafica la representación intermedia de la segunda capa.

```python
x1 = np.array([0,0,1,1]).reshape(-1,1)
x2 =  np.array([0,1,0,1]).reshape(-1,1)
X = np.hstack((x1,x2))
Y = np.array([[0],[1],[1],[0]])
lr = 0.001

for i in range(0,5):
    neural_network = NeuralNetworkXOR(X,Y,lr)
    neural_network.fit()

    y_rounded = np.around(neural_network.y_hat)

    colours = ListedColormap(['orange','b'])
    classes = list(map(str, np.abs(np.unique(y_rounded))))
    plot = plt.scatter(neural_network.h2_output[:,0],neural_network.h2_output[:, 1], c=y_rounded.reshape(1,-1)[0].tolist(),cmap=colours)

    plt.title("Experimento "+str(i+1)+" - Representación de Segunda Capa Oculta")
    plt.legend(handles=plot.legend_elements()[0],labels=classes)
    plt.grid(True)
    plt.show()
```

En los 5 experimentos podemos ver que en la representación de la capa intermedia es posible separar de manera lineal las 2 diferentes salidas de la función XOR, aún cuando cada representación intermedia nos ha devuelto diferentes valores en cada experimento.

## Conclusiones
- Backpropagation es un método que nos sirve para entrenar redes neuronales, modificando el peso de las neuronas por medio de la propagación del error en la neurona de salida.

- A pesar de tener diferentes valores en la representación de la capa intermedia, se ha logrado obtener el resultado esperado en la salida de la red neuronal.

- Por medio de entrenamiento de redes neuronales es posible romper la linealidad de una función.
